Creating a new feature in the dataset

```python
# Creating Car Age and Fuel Ranges features in the dataset
current_year = 2024

# Adding Car Age feature
E_vehicle_data['Car Age'] = current_year - E_vehicle_data['Model Year']

# Defining bins and labels for Fuel ranges
bins = [0, 100, 200, 300, float('inf')]
labels = ['Short Range', 'Medium Range', 'Long Range', 'Very Long Range']

# Adding Fuel Efficiency feature
E_vehicle_data['Fuel Range'] = pd.cut(E_vehicle_data['Electric Range'], bins=bins, labels=labels, right=False)

# Displaying the updated dataset structure to confirm changes
E_vehicle_data.info(), E_vehicle_data[['Car Age', 'Fuel Range']].head()
```

Selecting the feature and taget variable

```python
# Select features and target variable
features = ['County', 'City', 'Model Year', 'Make', 'Model',
            'Clean Alternative Fuel Vehicle (CAFV) Eligibility',
            'Electric Range', 'Base MSRP', 'Legislative District',
            'Electric Utility', 'Car Age', 'Fuel Range']
target = 'Electric Vehicle Type'

# Extracting features (X) and target (y)
X_features = E_vehicle_data[features]
y_target = E_vehicle_data[target]
```

Class Balancing

```python
# Importing necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from imblearn.over_sampling import SMOTE

# Encoding categorical features
cat_features = ['County', 'City', 'Electric Utility']
X_transformed = pd.get_dummies(X_features, columns=cat_features, drop_first=True)

# Label Encoding for other categorical features
label_encoder_make = LabelEncoder()
X_transformed['Make'] = label_encoder_make.fit_transform(X_transformed['Make'])

label_encoder_model = LabelEncoder()
X_transformed['Model'] = label_encoder_model.fit_transform(X_transformed['Model'])

label_encoder_cafv = LabelEncoder()
X_transformed['Clean Alternative Fuel Vehicle (CAFV) Eligibility'] = label_encoder_cafv.fit_transform(
    X_transformed['Clean Alternative Fuel Vehicle (CAFV) Eligibility']
)

label_encoder_fuel = LabelEncoder()
X_transformed['Fuel Range'] = label_encoder_fuel.fit_transform(X_transformed['Fuel Range'])

label_encoder_year = LabelEncoder()
X_transformed['Model Year'] = label_encoder_year.fit_transform(X_transformed['Model Year'])

# Handling imbalances using SMOTE
smote_balancer = SMOTE(random_state=42)
X_balanced, y_balanced = smote_balancer.fit_resample(X_transformed, y_target)
label_encoder = LabelEncoder()
y_numeric = label_encoder.fit_transform(y_balanced)
```

Splitting the data

```python
# Splitting data into training and test sets
X_train_balanced, X_test_balanced, y_train_balanced, y_test_balanced = train_test_split(
    X_balanced, y_numeric, test_size=0.2, random_state=42
)

# Scaling numerical features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_balanced)
X_test_scaled = scaler.transform(X_test_balanced)

# Confirming data shapes
{
    "X_train_shape": X_train_scaled.shape,
    "X_test_shape": X_test_scaled.shape,
    "y_train_shape": y_train_balanced.shape,
    "y_test_shape": y_test_balanced.shape,
}
```

Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Initializing Random Forest Classifier
rf_classifier = RandomForestClassifier(random_state=42)

# Training the model
rf_classifier.fit(X_train_scaled, y_train_balanced)

# Predicting on the test set
rf_y_pred = rf_classifier.predict(X_test_scaled)

# Evaluate the model
print("Random Forest Classifier Performance:")
print("Accuracy:", accuracy_score(y_test_balanced, rf_y_pred))
print("\nClassification Report:\n", classification_report(y_test_balanced, rf_y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test_balanced, rf_y_pred))
```

XGBOOST MODEL

```python
from sklearn.ensemble import GradientBoostingClassifier

# Initializing Gradient Boosting Classifier
gb_classifier = GradientBoostingClassifier(random_state=42)

# Training the model
gb_classifier.fit(X_train_scaled, y_train_balanced)

# Predicting on the test set
gb_y_pred = gb_classifier.predict(X_test_scaled)

# Evaluating the model
print("\nGradient Boosting Classifier Performance:")
print("Accuracy:", accuracy_score(y_test_balanced, gb_y_pred))
print("\nClassification Report:\n", classification_report(y_test_balanced, gb_y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test_balanced, gb_y_pred))
```

Plotting ROC curve for Random forest and XGBOOST

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
# Random Forest
rf_probabilities = rf_classifier.predict_proba(X_test_scaled)[:, 1]
rf_fpr, rf_tpr, _ = roc_curve(y_test_balanced, rf_probabilities)
rf_auc = auc(rf_fpr, rf_tpr)

# Gradient Boosting
gb_probabilities = gb_classifier.predict_proba(X_test_scaled)[:, 1]
gb_fpr, gb_tpr, _ = roc_curve(y_test_balanced, gb_probabilities)
gb_auc = auc(gb_fpr, gb_tpr)

# Plotting ROC Curves
plt.figure(figsize=(8, 6))
plt.plot(rf_fpr, rf_tpr, label=f"Random Forest (AUC = {rf_auc:.2f})")
plt.plot(gb_fpr, gb_tpr, label=f"Gradient Boosting (AUC = {gb_auc:.2f})")
plt.plot([0, 1], [0, 1], 'k--', label="Random Guessing")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

```python
# Removing dominant features
dominant_features = [index for _, index in rf_top_features[:2]]
X_train_dropped = X_train_scaled[:, [i for i in range(X_train_scaled.shape[1]) if i not in dominant_features]]
X_test_dropped = X_test_scaled[:, [i for i in range(X_test_scaled.shape[1]) if i not in dominant_features]]

# Retraing the model
rf_classifier.fit(X_train_dropped, y_train_balanced)
rf_dropped_accuracy = rf_classifier.score(X_test_dropped, y_test_balanced)
print("Random Forest Accuracy without Top Features:", rf_dropped_accuracy)
```

Feed Forward Neural Network

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, Input

# Build the model
model = Sequential([
    Input(shape=(X_train_scaled.shape[1],)),  # Define the input shape seperately
    Dense(128, activation='relu'),  # First hidden layer
    BatchNormalization(),
    Dropout(0.3),  # Dropout layer to prevent overfitting
    Dense(64, activation='relu'),  # Second hidden layer
    BatchNormalization(),
    Dropout(0.3),
    Dense(32, activation='relu'),  # Third hidden layer
    Dense(1, activation='sigmoid')  # Output layer with sigmoid for binary classification
])

# Compile the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(
    X_train_scaled, y_train_balanced,  # Training features and labels
    validation_split=0.2,  # Use 20% of training data for validation
    epochs=15,  # Set number of epochs
    batch_size=32,  # Batch size
    verbose=1  # Displays training progress
)
```

```python
test_loss, test_accuracy = model.evaluate(X_test_scaled, y_test_balanced, verbose=0)
print(f"Test Accuracy: {test_accuracy:.4f}")
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# Predicting probabilities and convert them to binary predictions
y_pred_probs = model.predict(X_test_scaled)
y_pred = (y_pred_probs > 0.5).astype(int).flatten()

# Classification report and confusion matrix
print("Classification Report:\n", classification_report(y_test_balanced, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test_balanced, y_pred))
```
